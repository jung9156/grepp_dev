```python
tt.isna().sum()
```

```python
tt['Cabin'] = tt['Cabin'].fillna('C000')
#fillna로 채운 값을 tt['Cabin']에 할당 해줘야 채움값이 반영된다(혹은 inplace옵션 이용)
tt.head(3)
```

```python
tt['Age'] = tt['Age'].fillna(tt['Age'].mean())
tt['Embarked'] = tt['Embarked'].fillna('s')
tt.isna().sum()
```

```python
tt['Name_len'] = tt['Name'].apply(lambda x: len(x))
tt[['Name', 'Name_len']].head(5)
```

```python
tt['Child_Adult'] = tt['Age'].apply(lambda x: 'child' if x <= 15 else 'Adult')
tt[['Age', 'Child_Adult']].head(8)
```

```python
tt['Age_cat'] = tt['Age'].apply(lambda x: 'Child' if x <= 15 else ('Adult' if x <= 60 else 'Elderly'))
tt['Age_cat'].value_counts()
```

```python
def get_category(age):
    cat = ''
    if age <= 5: cat = 'Baby'
    elif age <= 12: cat = 'Child'
    elif age <= 18: cat = 'Teenage'
    elif age <= 25: cat = 'Student'
    elif age <= 35: cat = 'Young Adult'
    elif age <= 60: cat = 'Adult'
    else : cat = 'Elderly'
    
    return cat

tt['Age_cat'] = tt['Age'].apply(lambda x: get_category(x))
tt['Age_cat'].value_counts()
```
